```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
traditional_saliency = np.array([
    [3, 0, 1],
    [0, 0, 0],
    [0, 2, 4]
])

deconvnet_saliency = np.array([
    [0, 2, 0],
    [0, 0, 0],
    [1, 2, 0]
])

# Chessboard visualization using matplotlib
def plot_saliency(saliency_map, title):
    plt.figure(figsize=(5, 5))
    plt.imshow(saliency_map, cmap="hot", interpolation="nearest")
    plt.colorbar(label="Importance")
    plt.title(title)
    plt.xticks(range(3), ["A", "B", "C"])  # Simulating chess columns
    plt.yticks(range(3), ["3", "2", "1"])  # Simulating chess rows
    plt.gca().invert_yaxis()
    plt.show()

# Plot Traditional Backpropagation Saliency Map
plot_saliency(traditional_saliency, "Traditional Backpropagation Saliency Map")

# Plot Deconvnet Saliency Map
plot_saliency(deconvnet_saliency, "Deconvnet Saliency Map")
```
